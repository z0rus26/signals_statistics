# **Random Walk (Implementación usando Python)**

Introducción:

Un camino aleatorio (random walk en inglés) es un objeto matemático tambien conocido como proceso estocastico, este describe la trayectoria que de una sucesión de pasos aleatorios definidos dentro de un conjunto de numeros como los enteros.
Un ejemplo representativo de *RW*, es una RW que empieza en 0 y a cada paso se mueve +1 ó -1 con igual probabilidad. *[Subhajit Saha, Random Walk (Implementation in Python), 09-02-2018, en: https://www.geeksforgeeks.org/random-walk-implementation-python/ ]*

La ruta trazada por una molécula mientras viaja por un líquido o un gas, el camino que sigue un animal en su búsqueda de comida, el precio de una acción fluctuante y la situación financiera de un jugador pueden tratarse como una caminata aleatoria. El término caminata aleatoria fue introducido por Karl Pearson en 1905. 
*[Pearson, K. (1905). The Problem of the Random Walk. Nature. 72, 294]*

```
#Código de python para generar un rw de una dimensión
import random 
import numpy as np 
import matplotlib.pyplot as plt 


# Factores de probabilidad para ir de arriba abajo
prob = [0.05, 0.95] 

# posición inicial
start = 2
positions = [start] 

# creación de puntos aleatorios 
rr = np.random.random(1000) 
downp = rr < prob[0] 
upp = rr > prob[1] 

# generación del camino
for idownp, iupp in zip(downp, upp): 
	down = idownp and positions[-1] > 1
	up = iupp and positions[-1] < 4
	positions.append(positions[-1] - down + up) 

# Gráfica
plt.plot(positions) 
plt.show() 
```

Expandiendo la idea anterior a multiples dimensiones, el conjunto de puntos recorrios aleatoriamente presenta una interesante representación gráfica, es posible entonces encontrar un fractal discreto, un conjunto que exhibe una autosemejanza estocástica.  

La trayectoria de una RW representa la colección de puntos de llegada o visitados, sin tener en cuenta necesariamente cuando llego la RW al punto.
En una dimensión la trayectoria estará dada simplemente entre el rango producido por los puntos de altura máxima y mínima.

```
#Código de python para generar un rw de dos dimensiones
import numpy 
import pylab 
import random 

# numero de pasos
n = 100000

#Reserva de memoria para 2 vectores que van a contener las coordenadas X y Y
x = numpy.zeros(n) 
y = numpy.zeros(n) 

# for loop para rellenar las coordenadas del RW
for i in range(1, n): 
	val = random.randint(1, 4) 
	if val == 1: 
		x[i] = x[i - 1] + 1
		y[i] = y[i - 1] 
	elif val == 2: 
		x[i] = x[i - 1] - 1
		y[i] = y[i - 1] 
	elif val == 3: 
		x[i] = x[i - 1] 
		y[i] = y[i - 1] + 1
	else: 
		x[i] = x[i - 1] 
		y[i] = y[i - 1] - 1
	

# Graficando
pylab.title("Random Walk ($n = " + str(n) + "$ pasos)") 
pylab.plot(x, y) 
pylab.savefig("rand_walk"+str(n)+".png",bbox_inches="tight",dpi=600) 
pylab.show() 
```
